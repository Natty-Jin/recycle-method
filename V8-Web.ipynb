```python
root_dir = "Ans"

# data
data_img_dir = os.path.join(root_dir, "data")
data_label_dir = os.path.join(root_dir,"data")
# valid
valid_img_dir = os.path.join(root_dir, "valid")
valid_label_dir = os.path.join(root_dir,"valid")
# test
test_img_dir = os.path.join(root_dir, "test")
test_label_dir = os.path.join(root_dir,"test")
```

```python
data = glob.glob(os.path.join(data_img_dir,"*.jpg"))
valid = glob.glob(os.path.join(valid_img_dir,"*.jpg"))
test = glob.glob(os.path.join(test_img_dir,"*.jpg"))
```

```python
import random

# data 리스트 내부의 인덱스를 랜덤으로 섞음
random.shuffle(data)
random.shuffle(valid)
random.shuffle(test)
```

```python
train = data[: 20000]
valid = valid[: 2000]
test = test[: 4000]
```

```python
# train.txt
with open(os.path.join(root_dir, "train.txt"), 'w') as f:
	f.write('\n'.join(train) + '\n')

# valid.txt
with open(os.path.join(root_dir, "valid.txt"), 'w') as f:
	f.write('\n'.join(valid) + '\n')

# test.txt
with open(os.path.join(root_dir, "test.txt"), 'w') as f:
	f.write('\n'.join(test) + '\n')
```

```python
import yaml

yaml_data = {"names":["PET_No_Label", 'PET_Label', 'PS_No_Label', 'PS_Label', 'PP_No_Label', 'PP_Label', 'PE_No_Label', 'PE_Label'], # 클래스 이름
             "nc":8, # 클래스 수

# yaml_data = {"names" : ["No_Label", "Label"],
#              "nc" : 2,
             "path":root_dir, # root 경로
             "train":os.path.join(r"C:\Users\Admin\Downloads\137 original data\01 data\Ans\train.txt"), # train.txt 경로
             "val":os.path.join(r"C:\Users\Admin\Downloads\137 original data\01 data\Ans\valid.txt"), # valid.txt 경로
             "test":os.path.join(r"C:\Users\Admin\Downloads\137 original data\01 data\Ans\test.txt") # test.txt 경로
             }

with open(os.path.join(root_dir, "custom.yaml"), "w") as f:
  yaml.dump(yaml_data, f)
```

```python
from ultralytics import YOLO

model = YOLO('yolov10b.pt')
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
model.train(data="Ans/custom.yaml", epochs= 100, patience= 30, batch= 8, imgsz= 640)  # train the model
```

```python
# Use the model
metrics = model.val()  # evaluate model performance on the validation set
results = model.predict(source= r"Ans\test/", save= True)  # predict on an image
path = model.export(format="onnx")  # export the model to ONNX format
```
